# Chapter 5 -Evaluate models - testdata

Now let’s refactor our model evaluation function to use Yellowbrick’s ClassificationReport class, a model visualizer that displays the precision, recall, and F1 scores. This visual model analysis tool integrates numerical scores as well as color-coded heatmaps in order to support easy interpretation and detection, particularly the nuances of Type I and Type II error, which are very relevant (lifesaving, even) to our use case!

Type I error (or a “false positive”) is detecting an effect that is not present (e.g. determining a mushroom is poisonous when it is in fact edible).

Type II error (or a “false negative”) is failing to detect an effect that is present (e.g. believing a mushroom is edible when it is in fact poisonous).

Confusion matrix (Geron, 2019)P.90

As for our purpose recall and percision are more important then accuracy
Accuracy is defined as the ratio of correctly predicted examples by the total examples.
FN results in a 

The precision is the proportion of relevant results in the list of all returned search results. 
The recall is the ratio of the relevant results returned by the search engine to the total number of the relevant results that could have been returned.

We aim for a High Recall rate (close 100%),  a high Probability of Detection
improving the recall reducess the precision, leads to false positvies. Gernon,2019 p.93
And for comparing the classifiers we use the F1 score https://machinelearningmastery.com/tour-of-evaluation-metrics-for-imbalanced-classification/

```python
import os
import shutil
from sklearn.pipeline import Pipeline
from sklearn.model_selection import validation_curve
from sklearn.linear_model import Ridge
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from IPython.display import Image, Markdown
%matplotlib inline
```

```python
models = [
	'ResNet50',
	'ResNet50V2',
	'ResNet101V2',
    'ResNet152V2',
	'InceptionV3',
	'Xception',
	'VGG16',
	'VGG19',
	'MobileNet',
	'MobileNetV2',
	'EfficientNetB7',
	'InceptionResNetV2',
	'DenseNet121',
    'DenseNet169',
	'NASNetLarge',
]
```

```python
# retive data to be posted on github
base = '/home/labuser/deeplearning/thesis/solutions/compare-dnn/results'
for model in models:
    print('importing results from %s' % model)
    shutil.copyfile(os.path.join(base, model,model+'_classification_report.pkl'),os.path.join(os.getcwd(),'data',model+'_classification_report.pkl'))
    shutil.copyfile(os.path.join(base, model,model+'_scores.pkl'),os.path.join(os. getcwd(),'data',model+'_scores.pkl'))
    shutil.copyfile(os.path.join(base, model,model+'_training.log'),os.path.join(os. getcwd(),'data',model+'_training.log'))
    
```

# Part 1 - Classification reports

```python
classreport = {}
for model in models:
    print('importing results from %s' % model)
    classreport[model] = pd.read_pickle(os.path.join(os.getcwd(),'data',model+'_classification_report.pkl'))
```

```python
for model in models:
    display(Markdown('## '+model))
    plt.figure(figsize=(20,5))
    data = classreport[model]
    sns.heatmap(data.iloc[:-1, :].T, annot=True)
    plt.savefig(os.path.join(os.getcwd(),'data',model+'_classification_report.png'))
    plt.show()
    
```

# Beter ranking models - scores macro avg

https://towardsdatascience.com/multi-class-metrics-made-simple-part-ii-the-f1-score-ebe8b2c2ca1

```python
dfc = pd.DataFrame()
for model in models:
    print('importing results from %s' % model)
    dft = pd.read_pickle(os.path.join(os.getcwd(),'data',model+'_scores.pkl'))
    dfc = pd.concat([dfc,dft],ignore_index=True)
```

```python
dfc = dfc.round(decimals=4)
```

```python
dfc.sort_values('test_acc',ascending=False).to_csv(os.path.join(os.getcwd(),'data','all_models_acc.csv'), encoding='utf-8',index=False)
```

## Ranking test scores 

```python
dfc[dfc['test_acc'] > 0.8].sort_values('test_acc',ascending=False) # best scoring 
```

```python
dfc[dfc['test_acc'] > 0.8].sort_values('test_acc',ascending=False).to_csv(os.path.join(os.getcwd(),'data','top_models_acc.csv'), encoding='utf-8',index=False)
```

```python
dfc[dfc['test_acc'] < 0.8] # worst scoring 
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x="model", y="recall_score", data=dfc[dfc['test_acc'] > 0.8].sort_values('test_acc',ascending=False))
```
